```python
with open('day13-list.txt', 'r') as file:
    content = [item.replace('x,', '').replace('\n', '').split(',') for item in file.readlines()]

#Clean up the data
def content_cleanup(items):
    items_i = []
    items_j = []
    for i in items[1]:
        if i.isdigit:
            val_i = int(i)
            items_i.append(val_i)
    
    for j in items[0]:
        if j.isdigit:
            val_j = int(j)
            items_j.append(val_j)
    items_join = items_j + items_i
    return items_join

#Part 1 Solutions
first_minute = content_cleanup(content)[0]
bus_id = [item for item in content_cleanup(content)[1:]]
def bus_id_multiplied_by_wait_time(wait_time, bus_ids):
    id_and_time = {}
    for i in bus_ids:
        j = 0
        modulo = wait_time % i
        while modulo != 0:
            modulo = (wait_time + j) % i
            if modulo == 0:
                minutes = (wait_time + j)
                id_and_time[minutes] = i
                break
            j += 1
    return id_and_time

#Part 1 Answer
dict_ = bus_id_multiplied_by_wait_time(first_minute, bus_id)
min_value = min(dict_)
relevant_bus = dict_[min_value]
multiply_both = (min_value-first_minute) * relevant_bus
print(
    f'For part 1, time waited was {min_value}, bus id for that time is {relevant_bus}, multiplying result by the difference from initial time is {multiply_both}'
)
print(dict_)
    
```

```python
#Part 2, the answer was at the very last of the results
from math import gcd


def compute_lcm(a):
    lcm = a[0]
    for i in a[1:]:
        lcm = lcm * i // gcd(lcm, i)
    return lcm


def part_one(timestamp, ids):
    buses = [int(i) for i in ids if i != 'x']
    wait_times = []
    for bus in buses:
        prev_bus = timestamp // bus * bus
        wait_times.append(prev_bus + bus - timestamp)
    return min(wait_times) * buses[wait_times.index(min(wait_times))]


def part_two(ids):
    buses = []
    for bus in ids:
        if bus != 'x':
            buses.append(int(bus))
        else:
            buses.append('x')
    timestamp = 0
    matched_buses = [buses[0]]
    while True:
        timestamp += compute_lcm(matched_buses)
        print(timestamp)
        for i, bus in enumerate(buses):
            if bus != 'x':
                if (timestamp + i) % bus == 0:
                    if bus not in matched_buses:
                        matched_buses.append(bus)
        if len(matched_buses) == len(buses) - buses.count('x'):
            break

    return timestamp


with open("day13-list.txt") as f:
    inp = f.readlines()
    print(part_one(int(inp[0]), inp[1].split(',')))
    print(part_two(inp[1].split(',')))
```
