```python
#Day 1 first challange

with open('day2-list.txt', 'r') as file:
    array = []
    for items in file:
        num = 0
        while num < 1:
            array.append(items.strip('\n'))
            num += 1

def cleanData(array):
    dict_ = {}
    list_ = array.split()
    minMax = list_[0].split('-')
    
    dict_['min'] = int(minMax[0])
    dict_['max'] = int(minMax[1])
    
    #double [] next to each other indicates:
    #first [] is the index's order that you want to pick,
    #second [] is the item inside the first []'s index in whichever sequence that you inputted inside the second []
    
    dict_['char'] = list_[1][0]
    dict_['pass'] = list_[2]
    
    return dict_

#Make a dictionary in a form of list '[]',
#'for' function was to make the program read every item inside 'listOriginal' list
newDict_ = [cleanData(newList) for newList in array]

def verification(cleanedList):
    countChars = cleanedList['pass'].count(cleanedList['char'])
    verifPass = cleanedList['min'] <= countChars <= cleanedList['max']
    return verifPass

#filter(function, iterable)
#function is a set of statement that will act as a condition to filter your data
#iterable is your data, could be list, sets, tuples, or containers of any iterators
validPass = list(filter(verification, newDict_))
countValidPass = len(validPass)
print(countValidPass)
```

```python
#Day 2 second challange

with open('day2-list.txt', 'r') as file:
    listOriginal = []
    for items in file:
        num = 0
        while num < 1:
            listOriginal.append(items.strip('\n'))
            num += 1

def cleaning(array_a):
    dict_a = {}
    list_a = array_a.split()
    minMax = list_a[0].split('-')
    
    dict_a['min'] = int(minMax[0])
    dict_a['max'] = int(minMax[1])
    dict_a['char'] = list_a[1][0]
    dict_a['pass'] = list_a[2]
    return dict_a

#Make a dictionary in a form of list '[]',
#'for' function was to make the program read every item inside 'listOriginal' list
dictClean = [cleaning(newDict) for newDict in listOriginal]

def newValidation(array_b):
    firstPosition = array_b['min']
    charFirstPosition = array_b['pass'][firstPosition-1] == array_b['char']
    lastPosition = array_b['max']
    charLastPosition = array_b['pass'][lastPosition-1] == array_b['char']
    
    #'XOr' or 'Exclusive Or' (^) was a python 'or' operator
    #This exclusive operator used when we have two conditions that can't be both true
    validation = charFirstPosition ^ charLastPosition
    return validation

valid = list(filter(newValidation, dictClean))
number = len(valid)
print(number)
```
